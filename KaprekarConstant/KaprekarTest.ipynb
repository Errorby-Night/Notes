## Kaprekar's Constant

So what I will do here is take a 4 digit number and arrange the digits in descending and ascending order.
Then, subtract them. If the result is the original number then it is a Kaprekar's constant.

```python
def kaprekar():
    for i in range(1000, 10000):
        num = str(i)
        numlist = [x for x in num]
        numlist.sort()
        num1 = ''
        for n in numlist:
            num1 += n
        p = int(num1)
        numlist.sort(reverse=True)
        num2 = ''
        for n in numlist:
            num2 += n
        q = int(num2)
        if (q - p) == i:
            print(i)
kaprekar()
```

```python
# Plotting to show the number of turns it takes for a number to reach 6174 i.e., the Kaprekar Constant.
import matplotlib.pyplot as plt
#import numpy as np
def kaps(n):
    num = str(n)
    numlist = [x for x in num]
    numlist.sort()
    num1 = ''
    num2 = ''
    for x in numlist:
        num1 += x
    numlist.sort(reverse=True)
    for x in numlist:
        num2 += x
    return (int(num2)-int(num1))

def plotandstuff():
    x = []
    #y = [a for a in range(1001, 9999)]
    for i in range(1001, 9999):
        num = i
        c = 1
        while num != 6174:
            num = kaps(num)
            c = c+1
            if c > 200:
                #print(i)
                break
        x.append(c)
    #print(x)
    #nx = np.array(x)
    #ny = np.array(y)
    plt.plot(x)
    plt.xlim([1000, 10000])
    plt.ylim([0, 100])
    plt.xlabel("4 Digit Numbers")
    plt.ylabel("Turns it took to reach the Kaprekar's Constant")
    plt.show()

plotandstuff()
```
